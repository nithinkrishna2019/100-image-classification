# Image classification using CNN and Transfer learning architectures

```python
import os
import pandas as pd
import numpy as np
    
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
PATH = os.getcwd()
```

```python
DATA_PATH = os.path.join(PATH, 'data')
data_dir_list = os.listdir(DATA_PATH)
```

```python
print(data_dir_list)
```

```python
img_rows=224
img_cols=224
num_channel=3
```

```python
img_data_list=[]
classes_names_list=[]
labels_index={}
labels=[]
```

```python
num_epoch=10
batch_size=64
```

```python
import cv2
```

### Reading the data

```python
for dataset in data_dir_list:
    classes_names_list.append(dataset) 
    print ('Loading images from {} folder\n'.format(dataset)) 
    label_id = len(labels_index)
    labels_index[dataset] = label_id
    img_list=os.listdir(DATA_PATH+'/'+ dataset)
    for img in img_list:
        input_img=cv2.imread(DATA_PATH + '/'+ dataset + '/'+ img )
        input_img_resize=cv2.resize(input_img,(img_rows, img_cols))
        img_data_list.append(input_img_resize)
        labels.append(label_id)
```

```python
len(input_img_resize)
```

```python
type(img_data_list)
```

```python
labels
```

```python
len(img_data_list)
```

```python
num_classes = len(classes_names_list)
print(num_classes)
```

```python
img_data = np.array(img_data_list)
```

```python
img_data = img_data.astype('float32')
```

## scaling the image between 0 to 1

```python
img_data /= 255
```

## size and shape of the images

```python
print (img_data.shape)
```

```python
num_of_samples = img_data.shape[0]
input_shape = img_data[0].shape
```

```python
print(num_of_samples)
```

```python
print(input_shape)
```

```python
from keras.utils import to_categorical

labels = to_categorical(labels, num_classes)
```

## shuffling the image labels to boost the robustness during the training phase of the neural net

```python
from sklearn.utils import shuffle

X, Y = shuffle(img_data, labels, random_state=2)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
y_test.shape
```

## As we can see the number of labels in the images are 100

```python
y_train.shape
```

# CNN architecture          Model-1

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers import Conv2D, MaxPooling2D

from sklearn.model_selection import GridSearchCV
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=input_shape))
model.add(Conv2D(32, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.layers
```

```python
model.weights
```

```python
model.layers[0].get_weights()
```

```python
hist = model.fit(X_train, y_train, batch_size=batch_size, epochs=num_epoch, verbose=1, validation_data=(X_test, y_test))
```

# validation loss and accuracy in classification of the image

```python
score = model.evaluate(X_test, y_test, batch_size=batch_size)

print('Test Loss:', score[0])
print('Test Accuracy:', score[1])
```

```python
from sklearn.metrics import confusion_matrix

Y_pred = model.predict(X_test)
print(Y_pred)
```

```python
y_pred = np.argmax(Y_pred, axis=1)
print(y_pred)
```

```python
print(confusion_matrix(np.argmax(y_test, axis=1), y_pred))
```

# trasnfer learning

```python
from keras.layers import Input, Dense
```

```python
image_input = Input(shape=(img_rows, img_cols, num_channel))
```

```python
from keras.applications.vgg16 import VGG16

model = VGG16(input_tensor=image_input, include_top=True, weights='imagenet')
```

```python
model.summary()
```

```python
last_layer = model.get_layer('fc2').output
out = Dense(num_classes, activation='softmax', name='output')(last_layer)
```

```python
from keras.models import Model

custom_vgg_model = Model(image_input, out)
custom_vgg_model.summary()
```

```python
for layer in custom_vgg_model.layers[:-1]:
    layer.trainable = False
```

```python
custom_vgg_model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
custom_vgg_model.fit(X_train, y_train, batch_size=batch_size, epochs=num_epoch, verbose=1, validation_data=(X_test, y_test))
```

## validation loss and accuracy of the VGG -16 model

```python
(loss, accuracy) = custom_vgg_model.evaluate(X_test, y_test, batch_size=batch_size, verbose=1)

print("[INFO] loss={:.4f}, accuracy: {:.4f}%".format(loss, accuracy * 100))
```

```python
Y_train_pred = custom_vgg_model.predict(X_test)
```

```python
y_train_pred = np.argmax(Y_train_pred, axis=1)
print(y_train_pred)
```

```python
print(confusion_matrix(np.argmax(y_test, axis=1), y_train_pred))
```
